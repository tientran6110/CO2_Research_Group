```python
import pandas as pd
```

```python
dataFinal = pd.read_csv('../sites/AUCprtempregression3row.csv')
```

```python

    import numpy as np
    %matplotlib inline
    import matplotlib.pyplot as plt
    import seaborn as sns
    import warnings
    warnings.filterwarnings("ignore")
    from sklearn.linear_model import LinearRegression
    from IPython.display import HTML
    import io
    import base64
    from scipy.spatial import distance
    from scipy import stats
    import os
    import nbformat as nbf
    


    def condition1(row):
        return abs(row['Daily_NEE'] - row['Daily_RECO_temp']) > (row['Daily_NEE'] - row['Daily_RECO_temp'] / 2)

    def condition2(row, df):
        time_range = pd.date_range(start=row.name - pd.DateOffset(1), end=row.name + pd.DateOffset(1))
        time_range = df.index.intersection(time_range)
        return df.loc[time_range]['Daily_P'].any() > 0

    def condition3(row, df):
        time_range = pd.date_range(start=row.name - pd.DateOffset(1), end=row.name + pd.DateOffset(1))
        time_range = df.index.intersection(time_range)
        return df.loc[time_range]['Daily_EF'].diff().dropna().any() > 0
    
        

    if 'Time' in dataFinal.columns:
        dataFinal['Time'] = pd.to_datetime(dataFinal['Time'])
        dataFinal.set_index('Time', inplace=True)
        dataFinal['Year'] = dataFinal.index.year
    
    unique_years = dataFinal['Year'].unique()

    deleted_points = {}
    not_deleted_points = {}

    for year in unique_years:
        data_subset = dataFinal.loc[dataFinal['Year'] == year].copy()

        data_subset['cond1'] = data_subset.apply(condition1, axis=1)
        data_subset['cond2'] = data_subset.apply(condition2, df=data_subset, axis=1)
        data_subset['cond3'] = data_subset.apply(condition3, df=data_subset, axis=1)

        # Calculate the total number of conditions met
        data_subset['conditions_met'] = data_subset[['cond1', 'cond2', 'cond3']].sum(axis=1)
        # Choose 3 conditions, or we need to weight the importants of condition when we want to take out points
        filtered_data_subset = data_subset[data_subset['conditions_met'] >=2]
        not_filtered_data_subset = data_subset[data_subset['conditions_met'] <2]
        
        deleted_points[year] = not_filtered_data_subset
        not_deleted_points[year] = filtered_data_subset


        # Plotting
        fig, axs = plt.subplots(2, 1, figsize=(20, 20), sharex=True)
        axs[0].plot(data_subset.index, data_subset['Daily_NEE'], label='Daily_NEE')
        axs[0].plot(data_subset.index, data_subset['Daily_RECO_temp'], label='Daily_RECO_temp')
        axs[0].scatter(filtered_data_subset.index, filtered_data_subset['predicted_RECO'], color='red', label='predicted_RECO')
        axs[0].set_title(f'Daily_NEE & Daily_RECO_temp over Time for year {year}')
        axs[0].legend()
        
        axs[1].plot(data_subset.index, data_subset['Daily_EF'], label='Daily_EF')
        axs[1].bar(data_subset.index, data_subset['Daily_P'], color = 'red', label='Daily_P')
        axs[1].set_title(f'Daily_EF & Daily_P over Time for year {year}')
        axs[1].legend()

        plt.ylabel('Values')
        plt.xlabel('Time')

        plt.tight_layout()
        plt.show()
        
# Press Cell - Current Output - Toggle Output to disable scorrling output of graph. 

        
```
